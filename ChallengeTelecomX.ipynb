```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/main/TelecomX_Data.json"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df.head()
```

```python
df_flattened = pd.concat(
    [
        df.drop(['customer', 'phone', 'internet', 'account'], axis=1),
        df['customer'].apply(pd.Series),
        df['phone'].apply(pd.Series),
        df['internet'].apply(pd.Series),
        df['account'].apply(pd.Series)
    ],
    axis=1
)

df_flattened.head()
```

```python
df_flattened = pd.concat(
    [
        df_flattened.drop('Charges', axis=1),
        df_flattened['Charges'].apply(pd.Series)
    ],
    axis=1
)
df_flattened = df_flattened.rename(columns={'Monthly': 'MonthlyCharges', 'Total': 'TotalCharges'})

df_flattened['TotalCharges'] = pd.to_numeric(df_flattened['TotalCharges'], errors='coerce')

df_flattened.nunique()
```

```python
import unicodedata

def normalize_string(s):
    if isinstance(s, str):
        s = s.lower()
        s = unicodedata.normalize('NFKD', s)
        s = s.encode('ascii', 'ignore').decode('utf-8')
        s = s.strip()
    return s

for col in df_flattened.columns:
    if df_flattened[col].dtype == "object":
        df_flattened[col] = df_flattened[col].apply(normalize_string)
```

```python
mapping = {
    "PaymentMethod": {
        "electronic check": ["electronic check", "cheque eletronico"],
        "mailed check": ["mailed check", "cheque por correio"],
        "bank transfer (automatic)": [
            "bank transfer (automatic)",
            "transferencia bancaria (automatica)"
        ],
        "credit card": ["credit card", "cartao de credito"]
    },

    "Churn": {"yes": ["yes", "sim"],
        "no": ["no", "nao"]
    }
}
```

```python
def apply_mapping(series, dic):
    for canonical, variants in dic.items():
        for v in variants:
            series = series.replace(v, canonical)
    return series
```

```python
for col, dic in mapping.items():
    df_flattened[col] = apply_mapping(df_flattened[col], dic)
```

Relatrio Final  Anlise de Evaso de Clientes (Churn)

Introduo
Este relatrio tem como objetivo analisar o fenmeno de evaso de clientes (churn), identificando padres, comportamentos e caractersticas associadas aos cancelamentos. A anlise busca compreender quais fatores influenciam a deciso dos clientes de deixar o servio, permitindo que a empresa desenvolva estratgias de reteno mais eficazes.

Limpeza e Tratamento de Dados
Os dados foram importados e passaram por um processo de limpeza para garantir consistncia e qualidade. Foram tratadas inconsistncias em valores categricos, corrigidas formataes incorretas, convertidas colunas numricas para os tipos adequados e normalizadas variveis de texto. Linhas com valores ausentes foram ajustadas ou removidas, e colunas irrelevantes foram descartadas para otimizar a anlise.

Anlise Exploratria de Dados
A anlise exploratria investigou a distribuio da varivel Churn e sua relao com variveis categricas e numricas. Foram gerados grficos que mostraram como caractersticas como tipo de contrato, forma de pagamento, servios adicionais influenciam a taxa de cancelamento. Tambm foram analisadas variveis numricas, como tempo de contrato (tenure) e total gasto (TotalCharges), comparando clientes que cancelaram e os que permaneceram.

Concluses e Insights
A investigao revelou padres claros: certos perfis de clientes possuem maior probabilidade de cancelar o servio. Elementos como contratos mensais, pagamentos automticos via mtodos especficos e baixa fidelizao aparecem fortemente associados ao churn. Clientes com menor tempo de permanncia tendem a cancelar mais. Esses achados ajudam a identificar pontos de vulnerabilidade no modelo atual de reteno.

Recomendaes
Com base nos insights obtidos, recomenda-se desenvolver aes especficas para os grupos com maior propenso ao churn. Estratgias como descontos para aumento de fidelizao, melhoria nos servios mais associados a cancelamentos e reviso das modalidades de pagamento podem contribuir para reduzir a evaso. Alm disso, sugere-se implementar monitoramento contnuo para detectar sinais de risco antes que o cancelamento acontea.
